````python
import networkx as nx
from statsmodels.distributions.empirical_distribution import ECDF
import numpy as np
import matplotlib.pyplot as plt
````

````python
# creiamo il grafo vuoto, indiretto
Fb = nx.Graph()
````

## Analisi del grado dei nodi di una rete
Una volta caricato il grafo in memoria una delle proprietà che solitamente si analizzano è la __distribuzione del grado__, una delle misure di centralità in una network. 

Dal momento che il grafo non è orientato, per ora non distinguiamo tra grado in ingresso (in-degree) e grado in uscita (out-degree).

In NetworkX il metodo `degree()` calcola il grado di ogni nodo e restituisce una **DegreeView**, che passata sotto al costruttore dict mi restituisce un dictionary con chiave id del nodo e valore il grado del corrispettivo nodo.

````python
degree = dict(Fb.degree())
````

### ottenere il grado di un singolo nodo

````python
degree[23], Fb.degree()(23), Fb.degree(23)
````

### ottenere il grado di un sottoinsieme di nodi

````python
Fb.degree([23,1,63,193,963])
````

### ottenere il nodo con grafo massimo

````python
#ottengo la coppia nodo,grado
print(max(degree.items(), key = lambda x: x[1]))
#per ottenere il nodo che ha grado massimo
print(max(degree.items(), key = lambda x: x[1])[0])
````

````python
max(degree, key = degree.get)  # COOL
````

````python
from collections import Counter
ordered_degree = Counter(degree).most_common()
#ottengo la coppia nodo,grado
print(ordered_degree[0])
#per ottenere il nodo che ha grado massimo
print(ordered_degree[0][0])
````

### stats sul grado

Una volta ottenuto il campione (lista dei gradi) posso calcolare alcune statistiche:
* il grado medio
* il grado massimo
* il grado minimo
* la mediana del campione
* la moda del campione

Per calcolare queste statistiche introduciamo due librerie per il calcolo scientifico: __numpy__ e __scipy__.

````python
degree_list =list(degree.values())
````

````python
print(f'Grado medio: {np.mean(degree_list)}')
print(f'Standard deviation: {np.std(degree_list)}')
print(f'Median: {np.median(degree_list)}')
print(f'Min: {np.min(degree_list)}')
print(f'Max: {np.max(degree_list)}')
````

### La distribuzione del grado - PDF/CDF/CCDF
Dato il nostro campione il primo passo nella costruzione della varie funzioni che descrivono la distribuzione è contare quante volte un elemento occorre.

In questo caso possiamo utilizzare il metodo __histogram__ di numpy.

Vediamo un esempio del suo funzionamento ...

````python
test = [1,2,3,4,1,2,3,1,2,3,4,5,1,2,3,4,1,2,3,4,5,6,1,2,2,2,2,3,6,6,6,6,6,7]
counts, bins = np.histogram(test, bins=np.arange(min(test),max(test)+2))
````

````python
Counter(test)
````

````python
print(counts)
print(bins)
````

````python
plt.figure(figsize = (6,2))
plt.hist(bins[:-1], bins, weights=counts, color = 'tomato')
plt.xticks([x+0.5 for x in range(1,max(test)+1)],range(1,max(test)+1))
plt.yticks(range(10))
plt.grid(True)
plt.show()
````

````python
plt.figure(figsize = (6,2))
pdf_test = [c/len(test) for c in counts]
plt.plot(bins[:-1],pdf_test , color = 'tomato',marker='o', linestyle='--', ms=5)
#plt.xticks([x+0.5 for x in range(1,max(test)+1)],range(1,max(test)+1))
#plt.yticks(range(10))
plt.grid(True)
plt.show()
````

### Probability distribution function - PDF
Ora possiamo calcolare la PDF, i.e. $ P(K = k) = \frac{count(k)}{N}$

dove $N$ è il numero di elementi del campione.



````python
np.arange(min(degree_list),max(degree_list)+1)

count,bins = np.histogram(degree_list, bins=np.arange(min(degree_list),max(degree_list)+2))
````

````python
pdf_fb = count / Fb.order()
print(pdf_fb)
````

````python
fig = plt.figure(figsize=(8,4)) 
plt.plot(bins[:-1],pdf_fb, color='tomato',marker='o', linestyle='--', ms=5)
plt.xlabel('Grado',size=15, weight = 'bold')
plt.ylabel('P(X=x)',size=15, weight = 'bold')
````

### Cumulative Distribution Function - CDF
* La funzione di ripartizione è definita come:
$ F_X(x) = P(X\leq x) $
* Per calcolare la CDF possiamo utilizzare l'oggetto ECDF del modulo statsmodel

````python
cdf_function = ECDF(degree_list)
x = np.unique(degree_list)
y = cdf_function(x)

# Visualizzazione
fig_cdf_function = plt.figure(figsize=(8,4))
assi = fig_cdf_function.gca()
#assi.loglog(x,y,color = 'red', linestyle = '--', marker= 'o',ms = 16)
assi.plot(x,y,color = 'tomato', linestyle = '--', marker= 'o',ms = 5)
assi.set_xlabel('Grado',size=15, weight = 'bold')
assi.set_ylabel('P(X<=x)',size=15, weight = 'bold')
assi.set_xticks(range(0,1300,50),range(0,1300,50),size = 12, rotation = 90)
plt.grid() 
````

````python
np.percentile(degree_list,80)
````

### Complementary Cumulative Distribution Function - CCDF
* $ P(X > x) $
* Per calcolare la CCDF possiamo utilizzare ancora ECDF del modulo statsmodel, ma al momento del plot usiamo il complementare di ogni elemento (1-y)

````python
cdf_function = ECDF(degree_list)
x = np.unique(degree_list)
y = cdf_function(x)

# Visualizzazione
fig_cdf_function = plt.figure(figsize=(8,4))
assi = fig_cdf_function.gca()
assi.loglog(x,1-y,color = 'tomato', linestyle = '--', marker= 'o',ms = 5)
assi.set_xlabel('Grado',size=15, weight = 'bold')
assi.set_ylabel('P(X>x)',size=15, weight = 'bold')
````

### Identificare gli hub
Infine vogliamo identificare quali sono i nodi con più alto grado.

Il primo passo è definire una soglia per quantificare cosa intendiamo per 'alto grado'. Per questo possiamo utilizzare la CDF. In particolare fissato una $p \in [0,1]$ vogliamo trovare il valore del grado $k$ tale che $F_X(k)=p$. Questa operazione corrisponde alla definizione di __percentile__.

In Python posso calcolare i quantili utilizzando la funzione di Numpy **percentile**. In particolare fissiamo $p=0.99$

#### parentesi percentile
![title](./imgs/percentile.png)

````python
np.percentile(test,50)
````

````python
np.percentile(test,80)
````

perchè la parte decimale .4? 
se il parametro method non è specificato, di default si calcola come $$i + g = p/100 * (n -1) +1$$
il risultato è un numero decimale la cui parte intera viene usata come posizione del percentile, la parte decimale (g) è la parte decimale che viene aggiunta al percentile intero

````python
print(f"posizione: {int(34*0.80)}\n\
formula i+g: {0.80 * (34-1)+1} \n\
risultato: {np.percentile(test,80)}")
````

per un valore più intuitivo, si possono usare i metodi lower, midpoint e higher

````python
print(np.percentile(test,80, method= 'lower'),\
np.percentile(test,80, method= 'midpoint'),\
np.percentile(test,80, method= 'higher'))
````

````python
len([x for x in test if x<=5]),len([x for x in test if x<=5.5]),len([x for x in test if x<=6])
````

__i percentili possiamo vederli (anche se non sempre in modo preciso) anche dalla CDF__

````python
cdf_function = ECDF(test)
x = np.unique(test)
y = cdf_function(x)

# Visualizzazione
fig_cdf_function = plt.figure(figsize=(10,6))
assi = fig_cdf_function.gca()
#assi.loglog(x,y,color = 'red', linestyle = '--', marker= 'o',ms = 16)
assi.plot(x,y,color = 'tomato', linestyle = '--', marker= 'o',ms = 5)
assi.set_xlabel('Intero',size=15, weight = 'bold')
assi.set_ylabel('P(X<=x)',size=15, weight = 'bold')
plt.grid(True)
````

# Random network

## 1. Erdős-Rényi graph
L'algoritmo restituisce un grafo $G_{n,p}$ con un numero di nodi $n$ e archi scelti con probabilità $p$ tra i $\frac{n (n-1) }{2}$ possibile edge -indiretti- 



__scegliamo la density come probabilità, che equivale a grado medio sul numero di vicini che può avere ogni nodo__

````python
print(f"Grado medio/numero di nodi -1 == densità\n  Grado medio/numero di nodi -1: {np.mean(degree_list) / (Fb.order()-1)}\n                        densità: {density}")
````

````python
n_nodi = Fb.order()
random_graph = nx.fast_gnp_random_graph(n_nodi,density)
````

````python
print("Facebook")
print('--Number of nodes: {}'.format(Fb.order()))
print('--Number of links: {}'.format(Fb.size()))
print("Erdos-Renyi")
print('--Number of nodes: {}'.format(random_graph.order()))
print('--Number of links: {}'.format(random_graph.size()))
````

````python
grado_medio_random = np.mean(list(dict(random_graph.degree()).values()))
grado_medio_Fb = np.mean(list(dict(Fb.degree()).values()))
print("Facebook")
print(f'--grado medio: {grado_medio_Fb}')
print(f'--density: {nx.density(Fb)}')
print("Erdos-Renyi")
print(f'--grado medio: {grado_medio_random}')
print(f'--density: {nx.density(random_graph)}')
````

````python
cdf = ECDF(list(dict(Fb.degree()).values()))
x = np.unique(list(dict(Fb.degree()).values()))
y = cdf(x)

cdf_random = ECDF(list(dict(random_graph.degree()).values()))
x_random = np.unique(list(dict(random_graph.degree()).values()))
y_random = cdf_random(x_random)

fig_cdf_fb = plt.figure(figsize=(10,4))
axes = fig_cdf_fb.gca()
axes.set_xscale('log')
#axes.set_yscale('log')
axes.plot(x,y,marker='o',ms=8, linestyle='--', label = 'FB')
axes.plot(x_random,y_random,marker='+',ms=10, linestyle='--', label = 'ER')
axes.set_xlabel('Grado',size=15, weight = 'bold')
axes.set_ylabel('P(X<=x)',size=15, weight = 'bold')

plt.legend()
plt.show()
````

## random graph al variare di p

````python
random_graphs_varp = []
for n in range(1,55,5):
    p = n/100
    random_graphs_varp.append(nx.fast_gnp_random_graph(g.order(),p))
    
random_degrees_varp = []
for random in random_graphs_varp:
    random_degrees_varp.append(np.mean(list(dict(random.degree()).values())))

plt.figure(figsize = (10,4))
grado_medio_g = np.mean(list(dict(g.degree()).values()))
plt.plot(random_degrees_varp, '.--', ms = 10,lw =1, color = 'navy', alpha = 0.5, label = 'null degree ')
plt.plot([grado_medio_g]*len(random_degrees_varp), color = 'royalblue', label = 'real degree')
plt.xticks(range(len(random_degrees_varp)),[str(n/100) for n in range(1,55,5)])
plt.xlabel('Values of p', weight = 'bold')
plt.ylabel('Degree medio', weight = 'bold')
#plt.plot([np.mean(random_degrees_varp)]*20, color = 'navy', label = 'random degree')
plt.legend()
````
